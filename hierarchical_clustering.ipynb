```python
# hierarchical clustering --> group customers based on purchasing behaviour when you don't know the number of grouos beforehand 
# hierarchical clustering builds a tree (dendogram) of each clusters using
# agglomerative(bottom up) or divisive(top-down)
```

```python
import pandas as pd 
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
```

```python
# sample data
data ={
    'Income': [15,16,17,60,62,63],
    'Spending_Score': [39,81,6,55,47,50]
}
df = pd.DataFrame(data)
```

```python
# visualize dendogram
plt.figure(figsize =(8,4))
dendogram = sch.dendrogram(sch.linkage(df, method='ward'))
plt.title("Dendogram")
plt.xlabel("Customers")
plt.ylabel("Euclidean Distance")
plt.show()
# why dendrogram? --> to visually determine the number of clusters. cut the longest 
# vertical line that doesn't cross another - the number of cuts = optimal clusters
```

```python
cluster = AgglomerativeClustering(n_clusters=2, metric='euclidean', linkage='ward')
df['Cluster'] = cluster.fit_predict(df)
```

```python
plt.scatter(df['Income'], df['Spending_Score'], c=df['Cluster'], cmap='rainbow')
plt.title("Customer Segments")
plt.xlabel("Income")
plt.ylabel("Spending Score")
plt.show()
```
