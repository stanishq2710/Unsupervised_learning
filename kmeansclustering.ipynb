```python
# k-means clustering --> unsupervised clustering 
# group similar types of customers based on purchasing behavior
# when to use? --> customer segmentation, pattern discovery, grouping unlabeled data
import pandas as pd 
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import make_blobs
```

```python
#simulate 2D data with 3 clusters
x,y = make_blobs(n_samples=300, centers=3, cluster_std=0.60, random_state=42)

#optional: Scale features
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
```

```python
kmeans  = KMeans(n_clusters=3, random_state=42)
kmeans.fit(x_scaled)
# get predited cluster labels
y_kmeans = kmeans.predict(x_scaled)
```

```python
plt.scatter(x_scaled[:,0], x_scaled[:,1], c=y_kmeans, cmap='viridis')
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], s=200, c='red', marker= 'X')
plt.title("K-Means Clustering")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

```python
# inputs --> only features (no labels required)

# how to choose K? --> using elbow method
inertia = []
for k in range(1,10):
    km = KMeans(n_clusters =k, random_state=42)
    km.fit(x_scaled)
    inertia.append(km.inertia_)
# For each k, it calculates inertia (how far data points are from their cluster center).
plt.plot(range(1,10), inertia, marker='o')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method')
plt.show()
```
