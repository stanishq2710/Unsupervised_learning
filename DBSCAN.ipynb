```python
# DBSCAN groups together points that are closely packed (points with many nearby neighbors)
# and  marks points that lie alone in low density regions (outliers)
# density based spatial clustering of application woth noise --> DBSCAN
# key parameters -> eps: max dist between two samples to be considered neighbors
# min_samples:min no of points to form a dense region 
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons #to demonstrate clustering
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
#generate synthetic data
x,y =make_moons(n_samples=300, noise =0.5, random_state=42)
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
```

```python
dbscan = DBSCAN(eps=0.3, min_samples=5)
clusters = dbscan.fit_predict(x_scaled)
```

```python
plt.figure(figsize=(8, 5))
plt.scatter(x_scaled[:, 0], x_scaled[:, 1], c=clusters, cmap='plasma', s=50)
plt.title("DBSCAN Clustering")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```
